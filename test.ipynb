```python
# 4! = 4*3*2*1
# np.arange(1, n+1)  # array([1, 2, 3, 4])
def factorial(n):
  return np.prod(np.arange(1, n+1))

factorial(4)
```

```python
def factorialRec(n):
  if (n == 0 or n == 1):
    return 1
  return n * factorialRec(n-1)

factorialRec(4)
```

```python
# verify:
import math

n = 4
print(math.factorial(n) == factorialRec(n))
```

```python
import numpy as np

def dotProduct (arr1, arr2):
  if isinstance(arr1, np.ndarray) and isinstance(arr2, np.ndarray) and len(arr1) == len(arr2):
    return sum(arr1 * arr2)
  raise Exception("wrong input")

a1 = np.array([1, 2, 3])
a2 = np.array([-1, -2, -3])
print(dotProduct(a1, a2))
```

```python
def fibonacciSeq(n):
  fib = [0, 1]
  for i in range(n-2):
    fib.append(fib[-2] + fib[-1])
  return fib

print(fibonacciSeq(6))
```

```python
import time
import numpy as np

toc = [0] * 3

# method 1: list append
tic = time.perf_counter()
N = 10000
lst = list()
for i in range(N):
  lst.append(i**2)
toc[0] = time.perf_counter() - tic

# method 2: numpy array append (least efficient, not recommended)
tic = time.perf_counter()
N = 10000
lst = np.array([]) # empty array
for i in range(N):
  np.append(lst, i**2)
toc[1] = time.perf_counter() - tic


# method 3: np.zero() initialization
tic = time.perf_counter()
lst = np.zeros(N)
for i in range(N):
  lst[i] = i**2
toc[2] = time.perf_counter() - tic

print(toc) # s
```

```python
import numpy as np

def poissonCounter(lam):
  target = np.exp(-lam) # lambda
  counter, currval = 0, 1
  while currval > target:
    currval *= np.random.rand() # uniform distribution over [0, 1)
    counter += 1
  return counter

for lam in range(10):
  print(poissonCounter(lam))
```

```python
def multiply2(x, y):
  if not isinstance(x, (int, float)):
    if isinstance(x, str) and x.isnumeric():
      print("converting x to float ...")
      x = float(x)
    else:
      raise Exception("invalid input")
  return x * y

print(multiply2(2, 3)) # 6
print(multiply2(2.0, 3)) # 6.0
print(multiply2("2", 3)) # 6.0
```

```python
def multiply3(x, y):
  if not isinstance(y, (int, float)):
    if isinstance(y, str) and y.isnumeric():
      y = float(y)
    else:
      raise Exception("invalid input")
  return x ** y

print(multiply3(2, 4))  # 16
print(multiply3(2, 2.5))  # 5.656854249492381
print(multiply3(2, "4"))  # 16.0
```

```python
def multiply4(x, y):
  try:
    return x ** y
  except:
    raise Exception("invalid input")

print(multiply4(2, 4))  # 16
print(multiply4(2, 2.5))  # 5.656854249492381
print(multiply4(2, "4"))  # Exception: invalid input
```

```python
import matplotlib.pyplot as plt # matplotlib is a dependency of seaborn
import seaborn as sns
import numpy as np
import pandas as pd

n = 200
D = np.zeros((n, 2))  # initialize a 200x2 matrix of zeros
D[:,0] = np.linspace(0, 10, n) + np.random.randn(n) # linear increase + noise
D[:,1] = D[:,0]**2 + np.random.randn(n)*10 # 1st col squared + noise

# plt.plot(D[:,0], D[:,1], "o")
sns.jointplot(x=D[:,0], y=D[:,1]) # jointplot = scatterplot + marginal histograms on x and y-axis
plt.show()
```

```python
# plotting with labels through pandas:
df = pd.DataFrame(data=D, columns=["varr1", "var2"])
sns.jointplot(x=df.columns[0], y=df.columns[1], data=df, kind="kde", color="purple") # kernel density estimation
plt.show()
```

```python
# create a regression plot of data in matrix D -> df
# sns.regplot(x=df.columns[0], y=df.columns[1], data=df)
sns.regplot(x=df.columns[0], y=df.columns[1], data=df, color="green")
plt.title(f"Regresson of {df.columns[0]} on {df.columns[1]}")
plt.show()
```
